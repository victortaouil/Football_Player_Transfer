```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
transferencias = pd.read_csv('top250-00-19.csv')
transferencias.sample(n=5)
```

```python
#Vamos retirar a coluna market_value por nao ter interesse nela nesta analise

transferencias = transferencias.drop(columns = 'Market_value')

# Vamos retirar o ';' da coluna transfer_fee

transferencias['Transfer_fee;'] = transferencias['Transfer_fee;'].str.strip(';')

# Vamos transformar as temporadas para que fique mais simples a interpretacao, transformaremos todas para o primeiro ano ( entao 2005 - 2006, ficara 2005)

transferencias['Season'].loc[transferencias['Season']  == '2000-2001'] = 2000
transferencias['Season'].loc[transferencias['Season']  == '2001-2002'] = 2001
transferencias['Season'].loc[transferencias['Season']  == '2002-2003'] = 2002
transferencias['Season'].loc[transferencias['Season']  == '2003-2004'] = 2003
transferencias['Season'].loc[transferencias['Season']  == '2004-2005'] = 2004
transferencias['Season'].loc[transferencias['Season'] == '2005-2006'] = 2005
transferencias['Season'].loc[transferencias['Season'] == '2006-2007'] = 2006
transferencias['Season'].loc[transferencias['Season'] == '2007-2008'] = 2007
transferencias['Season'].loc[transferencias['Season'] == '2008-2009'] = 2008
transferencias['Season'].loc[transferencias['Season'] == '2009-2010'] = 2009
transferencias['Season'].loc[transferencias['Season'] == '2010-2011'] = 2010
transferencias['Season'].loc[transferencias['Season'] == '2011-2012'] = 2011
transferencias['Season'].loc[transferencias['Season'] == '2012-2013'] = 2012
transferencias['Season'].loc[transferencias['Season'] == '2013-2014'] = 2013
transferencias['Season'].loc[transferencias['Season'] == '2014-2015'] = 2014
transferencias['Season'].loc[transferencias['Season'] == '2015-2016'] = 2015
transferencias['Season'].loc[transferencias['Season'] == '2016-2017'] = 2016
transferencias['Season'].loc[transferencias['Season'] == '2017-2018'] = 2017
transferencias['Season'].loc[transferencias['Season'] == '2018-2019'] = 2018





```

```python
#Vamos agora lidar com as colunas que nos informam sobre as ligas, por definicao iremos utilizar no df apenas as 20 ligas com maior numero de transferencias 
transferencias.sample(n=5)

criterio = transferencias['League_from'].value_counts()[:20]

criterio = pd.DataFrame(criterio)
criterio = criterio.reset_index()
criterio = criterio.rename(columns= {'index':'League_from','League_from':'Times'})
criterio.head(10)

# Agora Vamos fazer um inner join entre o criterio e as tranferencias.

transferencias = transferencias.merge(criterio, on= 'League_from')
transferencias.sample(n=10)

```

```python
#Agora vamos fazer a mesma coisa com a coluna 'League to', utilizar somente as 20 ligas que mais vezes compraram jogadores

criterio2 = transferencias['League_to'].value_counts()[:20]

criterio2 = pd.DataFrame(criterio2)
criterio2 = criterio2.reset_index()
criterio2 = criterio2.rename(columns= {'index':'League_to','League_to':'Times'})
criterio2.head(10)

#Agora vamos unir com o dataframe de transferencias 
transferencias2 = transferencias.merge(criterio2, on='League_to')

```

```python
#Nosso Dataframe agora se chama transferencias2. Vamos analisa-lo para ver se nao ha registros com erros
transferencias2.isnull().sum()
transferencias2 = transferencias2.drop(columns = ['Times_x','Times_y'])

transferencias2['Position'].unique()
transferencias2.isna().sum()

transferencias2['League_from'].loc[transferencias2['League_from'] == '1.Bundesliga'] = 'Bundesliga'
transferencias2['League_to'].loc[transferencias2['League_to'] == '1.Bundesliga'] = 'Bundesliga'
transferencias2['League_to'].unique()



```

```python
#Ultima checagem para ver se os registros estao em unidades corretas

transferencias2.info()
```
